## 4. Renaming columns and dropping columns

## 7. Handling missing or null values.

```python
null_values = cars_data[cars_data.isnull().any(axis = 1)]
null_values
```

Hp has null values. Hp will be filled by 0 as the cars having null values are electric cars. So they don't have Hp in general


```python
cars_data['hp'] = cars_data['hp'].fillna(0)
```

Fuel type null values will be fill with the mode found it before "regular unleaded" as this columns is a categorical one.

```python
cars_data['fuel_type'].mode()
cars_data['fuel_type'] = cars_data['fuel_type'].fillna('regular unleaded')
```

We know that electric cars doesn't have any cylinders so the null values will be again filled with 0

```python
cars_data['fuel_type'] = cars_data['fuel_type'].fillna(0)
```

 Doors will be fill with Mean (Integer) as it is a numerical value. And we can not have decimal numbers here.

```python
cars_data['doors'] = cars_data['doors'].fillna(cars_data['doors'].mean())
null_check = cars_data[cars_data.isnull().any(axis=1)]
null_check
```
