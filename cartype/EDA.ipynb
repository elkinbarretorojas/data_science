## 5. Handling missing or null values.

```python
cars_data.isnull().sum()
```

Hp has null values. Hp will be filled by 0 as the cars having null values are electric cars. So they don't have Hp in general


```python
cars_data.loc[cars_data['engine_hp'].isnull(),'engine_hp'] = cars_data['engine_hp'].fillna(0)
cars_data.head()
```

We know that electric cars doesn't have any cylinders so the null values will be again filled with 0

```python
engine_mode = cars_data['engine_cylinders'].mode()
```

```python
cars_data.loc[cars_data['engine_cylinders'].isnull(),'engine_cylinders'] = cars_data['engine_cylinders'].fillna(4)
```

Fuel type null values will be fill with the mode found it before "regular unleaded" as this columns are categoricals.

```python
cars_data.loc[cars_data['engine_fuel_type'].isnull(),'engine_fuel_type'] = cars_data['engine_fuel_type'].fillna('regular unleaded')
```

 Doors will be fill with Mean (Integer) as it is a numerical value. And we can not have decimal numbers here.

```python
cars_data.loc[cars_data['number_of_doors'].isnull(),'number_of_doors'] = cars_data['number_of_doors'].fillna(cars_data['number_of_doors'].mean())
```

```python
s1 = cars_data.shape
clean = cars_data[['year', 'engine_cylinders']]
for i in clean.columns:
    qt1 = cars_data[i].quantile(0.25)
    qt3 = cars_data[i].quantile(0.75)
    iqr =  qt3 - qt1
    lower = qt1-(1.5*iqr)
    upper = qt3+(1.5*iqr)
    min_in = cars_data[cars_data[i]<lower].index
    max_in = cars_data[cars_data[i]>upper].index
    cars_data.drop(min_in, inplace = True)
    cars_data.drop(max_in, inplace = True)
s2 = cars_data.shape
outliers = s1[0] - s2[0]
print("Deleted outliers are : ", outliers)
```

## 8. Test of Normality (Focus in highway_mpg to answer the initial question)

Our initial question was. What is the most Fuel efficient (MPG) type of car? In order to treat the data, we are gonna perform the normality tests.

```python
##Create a Histogram
plt.hist(cars_data['highway_mpg'], edgecolor='black',bins=20)
plt.show()
#From de Grpah we could be a Normal distribution data
```

```python
##Create a Q-Q pLot
sm.qqplot(cars_data['highway_mpg'], line='45')
plt.show()
#We can see the data is not normal distributed
```

```python
##Perform Shapiro-Wilk Test
print(shapiro(cars_data['highway_mpg']))
#As the P-value is less than 0.05. Then we reject the null hypothesis since we dont have sufficient evidence to say that sample does not come from a normal distribution
```

```python
##Perform Kolmogorov Test
print(kstest(cars_data['highway_mpg'], 'norm'))
#Since the p-value is less than .05, we reject the null hypothesis of the Kolmogorov-Smirnov test. This means we have sufficient evidence to say that the sample data does not come from a normal distribution.
```

To conclude, the variable Highway MPG has small deviations from a perfect normal distribution that would not prevent us from using parametric or non-parametric statistical methods.
